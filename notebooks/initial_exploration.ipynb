# Munging data

###### Data is dirty and needs cleaning up. Seems most sheets have some text at the top, which pandas is trying to read and generating lots of NaNs. 

* First objective should be do get rid of all text that isn't an axis name or data point
* Second objective to organise remaining axis names so we have the original index

### Cause of Death

```python
df = dataframes['cause_of_death']['Country-specific Estimates']
```

```python
df.dropna(how='all', inplace=True)
```

```python
NaN_mean = df.isnull().sum(axis=1)[:25].mean()
df = df.drop(df[df.isnull().sum(axis=1) > NaN_mean].iloc[:-2].index).reset_index(drop=True)
```

```python
df_name = df.iloc[0, 0]
df_name = re.sub('\W+', '_', df_name).lower()
df = df.drop(0)
```

```python
column_headers = df.iloc[:2].fillna(method='ffill', axis=1).fillna(method='ffill', axis=0)
column_headers = [(i, j) for i, j in zip(
    column_headers.values[0], column_headers.values[1])]
df.columns = pd.MultiIndex.from_tuples(column_headers)
df = df.drop([1, 2]).reset_index(drop=True)
```

```python
df.to_csv(data_dir_path + '/' + df_name + '.csv')
```
